# Najbolj popularne knjige na Book depository

V projektno nalogo sem zajela podatke 1000 najbol prodajanih knjig na spletni strani http://www.bookdepository.com/bestsellers?format=1,2&searchLang=123. Omejila sem se zgolj na knjiged v angleškem jeziku, iz seznama sem izvzela eletronske knjige, ter turistične vodiče. Zbirala sem podatke o uvrstitvi, avtorju, naslovu, mesecu izdaje, letu izdaje ter formatu knjige.

Za začetek si poglejmo tabelo z vsemi zbranimi podatki

```python
knjige=pd.read_csv('podatki.csv')
knjige
```

Poglejmo si v katerem letu je bilo izdanih največ knjig iz seznama. Podatki kažejo, da so trenutno najpopularnejše knjige izdane po letu 2010, kar ni presenetljivo.

```python
#v katerem letu je bilo izdanih največ knjig iz seznama
knjige_po_letih=knjige.groupby('leto_izdaje').size().plot()
```

Naslednji graf prikazuje 20 najpogostejših avtorjev na seznamu. Očitno je najbolša avtorjica J.K.Rowlong, ki je izdala zbirko sedmih knjig o mladem čarovniku Harry Pottru.

```python
#najboljših 20 avtorjev
avtor=knjige['avtor'].value_counts().head(20)
avtor
```

Oglejmo si knjige, ki pripadajo najboljši avtorici:

```python
#razpredelnica knjig najpogostejše avtorice
najpogostejsa_avtorica=knjige[knjige['avtor'] == "J. K. Rowling"]
najpogostejsa_avtorica
```

```python
najpogostejsa_avtorica[['naslov', 'leto_izdaje']].sort_values('leto_izdaje')
```

Kot vidimo je saga o Harry Pottru res popularna, saj je avtorica po njen izdala še nekaj knjig, ki pa očitno niso tako uspešne.

Poglejmo si še kater format je bolj priljubljen. Kot vidimo so knjige z mehkimi platnicami veliko bolj popularne, kar spet ni presenetljivo ker so te knjige po navadi lažje za izdelavo ter zato cenejše pa tudi bolj praktične. V nadaljevanju si bomo podrobneje poglejmo nekaj podatkov o vsakem formatu.

```python
#delež knjig s trdimi in mehkimi platnicami
vrsta_platnice=knjige['format'].value_counts().plot(kind='pie')
```

Spodnja tabela prikazuje knjige s trdimi platnicami urejene glede na leto izdaje. Vidimo zanimiv podatek, da je bila najstarejša knjiga iz seznama izdana šele leta 2001. Ta podatek se mi zdi presnetljiv, bi pričakovala, da bodo starejše knjige izdane v trdi vezavi, novejše pa v mehki.


```python
#knjige s trdimi platnicami 
trde_platnice=knjige[knjige['format'] == "Hardback"].sort_values('leto_izdaje')
trde_platnice
```

Pa poglejmo še mehke platnice. Glede na to, da je knjig z mehkimi platnicami veliko več, me je zanimalo, ali ima tudi tukaj J.K. Rowling takšno prednost pred ostalimi. Očitno tukaj prevladujeta Julia Donaldson, avtorica otroških knjig ter Sarah J. Maas, ki izdaja fantazijske knjige.

```python
#najboljši avtorji knjig z mehkimi platnicami
mehke_platnice=knjige[knjige['format'] == "Paperback"]
avtorji_mehke_platnce=mehke_platnice['avtor'].value_counts().head(20).plot.bar()
```

Pred nami je graf s povprečnimi uvrstitvami knjig z mehkimi platnicami po letih. Zanimalo me je, ali so knjige iz določenega obdobja še visoko uvrščene, vendar graf prikazuje, da nobeno leto posebno ne izstopa.

```python
#povprečna uvrstitev knjig z mehkimi platnicami
povprecna_uvrstitev=mehke_platnice.groupby('leto_izdaje')
povprecna_uvrstitev['uvrstitev'].mean()
```

Pa pojdimo še malo v zgodovino. Spodaj imamo tabelo s knjigami, ki so nastale pred letom 2000 razvrščene glede na uvrstitev. Vidimo da jih ni ravno veliko. Na prvem mestu se znajde vsem dobro poznana knjiga Mali princ

```python
#razpredelnica knjig izdanih pred letom 2000
knjige_20_stoletja=knjige[knjige.leto_izdaje < 2000]
knjige_20_stoletja.sort_values('uvrstitev')
```

```python
Sedaj pa poglejmo še kinjige zadnjih nekaj let.
```

```python
#razpredelnica knjig izdanih po letu 2015
knjige_2015=knjige[knjige.leto_izdaje > 2015]
knjige_2015.sort_values('uvrstitev')
```
